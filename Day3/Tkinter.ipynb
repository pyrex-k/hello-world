# Tkinter

```python
import tkinter as tk
root = tk.Tk()
```

```python
root.title("My First Tkinter Window")
```

```python
root.mainloop() # the asterisc means it is busy
```

```python
welcome_label  = tk.Label(root, text = "Welcome to Tkinter") # execute this witout executing 
```

```python
welcome_label.pack() #pckages the thing in the application we are building
```

```python
# prepare some code for when the users click on the buttons
```

```python
def on_button_click():
    welcome_label.config(text="Button Clicked!") 
#when we clock on the button, the button should we change the text we have on the label
```

```python
click_me_button = tk.Button(root, text="Click Me!", command=on_button_click)
```

```python
# now pack the button 
```

```python
click_me_button.pack()
```
