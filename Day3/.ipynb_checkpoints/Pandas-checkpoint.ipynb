```python
df2.pivot_table(
    index = "rooms",
    aggfunc = "count",
    values = "city")
# i want to count how many city there are for each value of room
```

```python
# now plot it 
# associate pivot table with a variable and then decide the plot to use -> bar 
```

```python
pivot = df2.pivot_table(
    index = "rooms",
    aggfunc = "count",
    values = "city")
```

```python
pivot.plot(kind = "bar")
```

```python
max_price = df2["total (R$)"].max()
print(f"The maximum prices of the house is {max_price}")
```

```python
threshold = max_price * 0.9
print(f"The threshold is {treshold}")
```

```python
top_df = df2[df2["total (R$)"]>threshold]
```
