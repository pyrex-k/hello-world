```python
from mpl_toolkits.mplot3d import Axes3D

cathegories = ["Company A", "Company B", "Company C", "Company D"]

values = [20, 50, 10, 40]

colors = ["green", "blue", "red", "orange"]

fig = plt.figure() # preparing the container
ax = fig.add_subplot(111, projection="3d")
# grid 1 by 1 and then i'm plotting the element number 1: 111

x= np.arange(len(categories))

ax.bar(x,values, zdir='y', color=colors)

plt.bar(cathegories,values,color=colors)
plt.title("Bar Chart")
plt.xlabel("Northwest Companies")
plt.ylabel("Revenues")
plt.xticks(rotation=45)
plt.show()
```
