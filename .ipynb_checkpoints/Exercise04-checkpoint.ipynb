# FOR LOOPS

```python
range(5)
```

```python
for i in range(5):
    print(i)
```

```python
my_list = [1,4,7,10,-5]
my_list
```

```python
for i in my_list:
    print(i)
# the important thing to secure is an ordered list of elements (tuple, list)
```

```python
my_tuple=(3, -2, "hello", True)
my_tuple
```

```python
for i in my_tuple:
    print(i)
```

```python
for i in range(4):
    print("Barbera")
```

```python
# we can concatenate with a f string 
```

```python
for i in range(4):
    print(f"Barbera {i}")
```

# While

```python
counter=0
while counter < 5:
    print(counter)
    counter +=1
```

```python
counter=0
while counter < 5:
    print(counter)
    counter +=1
    if counter ==3:
        break
print(counter)
```

# Break

```python
counter=0
while counter < 5:
    print(f"within the loop: {counter}")
    counter +=1
    if counter ==3:
        break
print(f"Outside of the loop:{counter}")
```

# Error Handling

```python
try:
    x = 3/0 

except ZeroDivisionError as e: 
    print(e)
```

```python
try:
    x = 10 + "abc"

except TypeError as e:
    print(f"The following error occurred: {e}")
```

```python
try: 
    my_list = [1,2,3,4,5]
    print(my_list[10])
  
except IndexError as e:
    print(f"The following eorre tyoe has happenend {e}")
```
